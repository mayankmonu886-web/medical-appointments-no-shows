```python
df.columns = df.columns.str.replace('-', '_').str.strip()
df = df.rename(columns={'Handcap': 'Handicap'})
```

```python
df = df.drop(columns=['PatientId', 'AppointmentID'])
```

```python
df = df[(df['Age'] >= 0) & (df['Age'] <= 100)]
```

```python
print(f"Cleaned dataset size: {df.shape[0]} rows")
print("\nFirst 3 rows of your clean data:")
print(df.head(3))
```

```python
# 1. Calculate the exact number of days between scheduling and the appointment
# We subtract the booking day from the appointment day, then extract the '.days'
['WaitTime'] df= (df['AppointmentDay'] - df['ScheduledDay'].dt.normalize()).dt.days

# 2. Fix the "Time Travel" data error
# An appointment cannot physically happen BEFORE it was booked. 
# If WaitTime is negative (less than 0), it is a system logging error. We drop those rows!
df = df[df['WaitTime'] >= 0]

# 3. Create a second new feature: Day of the Week
# This extracts the name of the day (e.g., 'Monday', 'Friday') from the appointment date
df['DayOfWeek'] = df['AppointmentDay'].dt.day_name()

# 4. Check our brand new columns!
print("Success! New features created.")
print(df[['WaitTime', 'DayOfWeek']].head(4))
```

```python
# Save your updated data to a new file name so you don't overwrite your original raw data
df.to_csv('cleaned_medical_data.csv', index=False)

print("Success! Your clean dataset with the new columns has been saved as 'cleaned_medical_data.csv'.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean, professional look for your graph
sns.set_theme(style="whitegrid")

# Create the bar chart frame
plt.figure(figsize=(7, 5))

# Plot 'No_show' on the bottom axis and your new 'WaitTime' on the vertical axis
# Note: Since 'No_show' was mapped to 0 and 1, we can label them manually for clarity
sns.barplot(x='No-show', y='WaitTime', data=df, errorbar=None, palette='Set2')

# Add titles and labels so your manager can read it easily
plt.title('Impact of Wait Time on Appointment No-Shows', fontsize=14, pad=15)
plt.xlabel('Patient Attendance Status', fontsize=12)
plt.ylabel('Average Wait Time (Days)', fontsize=12)

# Change the 0 and 1 on the bottom axis to actual descriptive words
plt.xticks([0, 1], ['Showed Up (0)', 'No-Show / Missed (1)'])

# Save the actual image graph to your computer as a PNG picture
plt.savefig('wait_time_vs_noshow.png', dpi=300, bbox_inches='tight')

# Display the final chart on your screen
plt.show()
```
